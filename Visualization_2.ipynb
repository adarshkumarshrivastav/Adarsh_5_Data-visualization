```python
#Q.4:- Using the given x and y data,generate a 3D surface plot to visualize the function z = sin(sqrt( x**2 + y**2))
#      x = np.linspace(-5,5,100)
#      Y = np.linspace(-5,5,100)
#      x,y = np.meshgrid(x,y)
#      z = np.sin(np.sqrt(x**2 + y**2))
#       data = {'x':x.flatten(),'y':y.flatten(),'Z':z.flatten()}
#        df = pd.DataFrame(data)
# Ans:-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays and create a DataFrame (optional)
data = {'x': x.flatten(), 'y': y.flatten(), 'z': z.flatten()}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Add a color bar which maps values to colors
fig.colorbar(surf, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
#Q.5:-Using the given dataset,create a bubble chart to represent each country's population(y-axis),GDP(x-axis),and bubble size proporational to the population.
#  np.random.seed(25)
#  data = {'Country':-['USA','Canada','UK','Germany','France'],'Population':
#    np.random.randint(100,1000,5),'GDP':np.random.randint(500,2000,5)}
#     df  = pd.DataFrame(data)
#Ans:-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate the dataset
np.random.seed(25)  # For reproducibility
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(
    df['GDP'], df['Population'],
    s=df['Population'] * 10,  # Bubble size proportional to population
    alpha=0.6,  # Transparency
    c='blue',  # Color of bubbles
    edgecolors='w',  # Edge color of bubbles
    linewidth=0.5
)

# Add labels and title
for i, row in df.iterrows():
    plt.text(row['GDP'], row['Population'], row['Country'],
             fontsize=9, ha='center', va='center')

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Country Population vs. GDP')

# Show the plot
plt.show()
```

```python
#Q.1:-Create a Bokeh plot dispalying a sine wave.Set x-values from 0 to 10 and y-values as the sine of x.
#Ans:-
from bokeh.plotting import figure, show, output_file
import numpy as np

# Generate the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Set up the output file (you can also use output_notebook() for inline display in Jupyter)
output_file("sine_wave.html")

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add a line renderer with the data
p.line(x, y, legend_label="Sine Wave", line_width=2, color='blue')

# Show the plot
show(p)
```

```python
#Q.2:- Create a Bokeh scatter plot using randomly generated x and y values.Use different sizes and colors for the markers based on the 'sizes' and 'colors'columns.
#Ans:-
from bokeh.plotting import figure, show, output_file
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.rand(n) * 50 + 10  # Marker sizes
colors = np.random.rand(n)  # Marker colors

# Create a DataFrame (optional)
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Set up the output file
output_file("scatter_plot.html")

# Create a new plot
p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add a scatter renderer with different sizes and colors
p.scatter(
    x='x',
    y='y',
    size='sizes',
    color='colors',
    source=df,
    fill_alpha=0.6,
    legend_label="Data Points"
)

# Add labels and title
p.title.text_font_size = '16pt'
p.xaxis.axis_label_standoff = 12
p.yaxis.axis_label_standoff = 12
p.legend.label_text_font_size = '12pt'

# Show the plot
show(p)
```

```python
#Q.4:- Create a Bokeh histogram to visualize the distribution of the given data.
#       data_hist = np.random.randn(1000)
#       hist, edges  = np.histogram(data_hist,bins=30)
from bokeh.plotting import figure, show, output_file
import numpy as np

# Generate the data
np.random.seed(42)  # For reproducibility
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Set up the output file
output_file("histogram.html")

# Create a new plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', plot_height=400, plot_width=600)

# Add a quad renderer for the histogram
p.quad(
    top=hist,
    bottom=0,
    left=edges[:-1],
    right=edges[1:],
    fill_color='navy',
    line_color='white',
    alpha=0.7
)

# Customize the appearance
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

# Show the plot
show(p)
```

```python
# Q.5:- Create a Bokeh heatmap using the provided dataset.
#     data_heatmap = np.random.rand(10,10)
#     x=np.linspaces(0,1,10)
#     y= np.linspace(0,1,10)
#      xx,yy = np.meshgrid(x,y)
#Ans:-
from bokeh.plotting import figure, show, output_file
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Generate the data
np.random.seed(42)  # For reproducibility
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for use with Bokeh
data_flat = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}
df = pd.DataFrame(data_flat)

# Set up the output file
output_file("heatmap.html")

# Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', plot_height=400, plot_width=400, 
           x_range=(0, 1), y_range=(0, 1))

# Define the color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=df['value'].min(), high=df['value'].max())

# Add the heatmap as a quad renderer
p.quad(
    top=df['y'] + (y[1] - y[0]),
    bottom=df['y'],
    left=df['x'],
    right=df['x'] + (x[1] - x[0]),
    line_color=None,
    fill_color=mapper,
    source=df
)

# Customize the appearance
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

# Add a color bar
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
