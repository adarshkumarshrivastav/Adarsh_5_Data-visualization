```python
#Q.4:-Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
#Ans:-
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C']
values_A = np.random.normal(loc=10, scale=2, size=100)
values_B = np.random.normal(loc=15, scale=3, size=100)
values_C = np.random.normal(loc=20, scale=2.5, size=100)

# Combine the data
data = [values_A, values_B, values_C]

# Create a box plot to visualize the distribution
plt.boxplot(data, labels=categories, patch_artist=True, 
            boxprops=dict(facecolor='lightblue'))

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Distribution of Numerical Values Across Categories')

# Show the plot
plt.show()
```

```python
#Q.5:- Generate a synthetic dataset with correlated features.Visualize the correlation matrix of a dataset using a heatmap.
#Ans:-
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)  # For reproducibility
n_samples = 100

# Create correlated features
x = np.random.rand(n_samples)
y = 0.5 * x + np.random.rand(n_samples) * 0.5  # y is correlated with x
z = 0.3 * x + 0.7 * y + np.random.rand(n_samples) * 0.3  # z is correlated with x and y

# Combine into a DataFrame
df = pd.DataFrame({'Feature X': x, 'Feature Y': y, 'Feature Z': z})

# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0,
            cbar_kws={'shrink': 0.8}, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Q.1:- Using the given dataset,to generate a 3D scatter plot to visualize the distribution of data points in a three-dimensional space.
#          np.random.speed(30)
#         data = { 'X': np.random.uniform(-10,10,300),'Y':np.random.uniform(-10,10,300),'Z': np.random.uniform(-10,10,300)}
#         df = pd.DataFrame(data)
#Ans:-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the dataset
np.random.seed(30)  # For reproducibility
data = { 
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot of Dataset')

# Show the plot
plt.show()
```

```python
#ANS:-  For 2 .(B)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the sales data
np.random.seed(20)  # For reproducibility
data_sales = { 
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

# Pivot the data to get a matrix format
pivot_table = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap to visualize the variation in sales across different months and days
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5)
plt.xlabel('Month')
plt.ylabel('Day')
plt.title('Variation in Sales Across Different Months and Days')
plt.show()
```
